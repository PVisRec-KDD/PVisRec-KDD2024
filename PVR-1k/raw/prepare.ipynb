```python
import pickle
from collections import defaultdict
import re
with open('./1k/tmp/wide-and-deep-used_variable_config_mapping_list.pkl','rb') as f:
	user2Vis=pickle.load(f)
# for vis in user2Vis:
	# if len(vis)<4:
	# print(vis)
dataset_list=[]
attr_list=[]
attr_dict=defaultdict(list)
used_pair_dict=defaultdict(list)
used_attr_dict=defaultdict(list)
used_cfg_dict=defaultdict(list)
cfg_list=[]
for vis in user2Vis:
	attr_list.append(vis[1])
	attr_list.append(vis[2])
	split1=vis[3].split("type', '")
	if split1 and len(split1)>1:
		split2=split1[1].split("'")
		dataset_list.append(vis[0])
		if vis[1] not in attr_dict[vis[0]]:
			attr_dict[vis[0]].append(vis[1])
		if vis[2] not in attr_dict[vis[0]]:
			attr_dict[vis[0]].append(vis[2])
		used_pair_dict[vis[0]].append((vis[1],vis[2],split2[0]))
		used_attr_dict[vis[0]].append((vis[1],vis[2]))
		used_cfg_dict[vis[0]].append(split2[0])
		cfg_list.append(split2[0])
cfg_unilist=list(set(cfg_list))
attr_dict=dict(attr_dict)
used_pair_dict=dict(used_pair_dict)
attr_dict
```

```python
with open('../user2Vis.pkl','wb') as f:
	pickle.dump(used_pair_dict,f)
with open('../user2AttrAll.pkl','wb') as f:
	pickle.dump(attr_dict,f)
with open('../configAll.pkl','wb') as f:
	pickle.dump(cfg_unilist,f)
with open('../userLikeATTR.pkl','wb') as f:
	pickle.dump(used_attr_dict,f)
with open('../userLikeCFG.pkl','wb') as f:
	pickle.dump(used_cfg_dict,f)
```
